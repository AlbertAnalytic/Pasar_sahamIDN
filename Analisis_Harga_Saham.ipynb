## **Analisis Performa Saham Berdasarkan Harga Terakhir dan Kapitalisasi Pasar**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/DaftarSaham.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.dropna(subset = ['LastPrice', 'MarketCap'], inplace=True)
```

```python
df.info()
```

```python
#analisis deskriptif untuk memahami distribusi harga sama dan kapitalisasi pasar
#Descriptiv
print(df[['LastPrice','MarketCap']].describe())

#Ditribusi Harga Saham
plt.figure(figsize=(10,6))
sns.histplot(df['LastPrice'], kde=True)
plt.title('Distribusi Harga Saham')
plt.xlabel('Harga Terakhir')
plt.ylabel('Frequensi')
plt.show()
```

```python
#Distribusi Kapitalisasi Pasar
plt.figure(figsize=(10,6))
sns.histplot(df['MarketCap'], kde=True)
plt.title('Distribusi Harga Saham')
plt.xlabel('Kapitalisasi Pasar')
plt.ylabel('Frequensi')
plt.show()
```

```python
#Visualisasi Hubungan antara harga saham dan Kapitalisasi Pasar
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LastPrice', y='MarketCap', data=df)
plt.title('Harga Saham vs Kapitalisasi Pasar')
plt.xlabel('Harga Terakhir')
plt.ylabel('Kapitalisasi Pasar')
plt.show()
```
